```python
import spacy
import json


nlp = spacy.load("en_core_web_sm")

def get_noun(sentence,nlp):
    doc = nlp(sentence)
    nouns = [token.text for token in doc if token.pos_ == "NOUN"]
    return nouns

def noun2class(nouns):
    cls_str = ""
    for noun in nouns:
        cls_str += f"a {noun}."

naration_data = json.load(open("/root/videollm-online/datasets/ego4d/v2/annotations/refined_narration_stream_val.json"))
for video_id, video_data in naration_data.items():
    for narrations in video_data.values():
        for naration in narrations:
            nouns = get_noun(naration['text'],nlp)
            print(f" sentence: {naration} \n nouns: {nouns} \n ")
        break



```
